```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))
data_ad = np.genfromtxt('fix-result0-AD.txt', names=('0','1', '2','3','4'))
data_nc = np.genfromtxt('fix-result0-NC.txt', names=('0','1', '2','3','4'))
all_data_AD = [data_ad[str(i)] for i in range(5)]
all_data_NC = [data_nc[str(i)] for i in range(5)]

# plot violin plot
axes[0].violinplot(all_data_AD,
                   showmeans=False,
                   showmedians=True)
axes[0].set_title('AD')

axes[1].violinplot(all_data_NC,
                   showmeans=False,
                   showmedians=True)
axes[1].set_title('nc')

for ax in axes:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data_AD))])
    ax.set_xlabel('c1')
    ax.set_ylabel('confidence')

# add x-tick labels
plt.setp(axes, xticks=[y + 1 for y in range(len(all_data_AD))],
         xticklabels=['-2', '-1', '0', '1', '2'])
plt.show()


fig2, axes2 = plt.subplots(nrows=1, ncols=2, figsize=(9, 4))

# plot violin plot
axes2[0].boxplot(all_data_AD)
axes2[0].set_title('AD')

axes2[1].boxplot(all_data_NC)
axes2[1].set_title('NC')

for ax in axes2:
    ax.yaxis.grid(True)
    ax.set_xticks([y + 1 for y in range(len(all_data_AD))])
    ax.set_xlabel('c1')
    ax.set_ylabel('confidence')

# add x-tick labels
plt.setp(axes, xticks=[y + 1 for y in range(len(all_data_AD))],
         xticklabels=['-2', '-1', '0', '1', '2'])
plt.show()
```
